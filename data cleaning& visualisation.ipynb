```python
# Changing the datatype of some columns later on doing for the visualisation 


# the dtype is in the object format i cannot use the str function 
# convert that to string first 
data['discounted_price'] = data['discounted_price'].astype(str)
```

```python
data['discounted_price'] = data['discounted_price'].str.replace('₹', '').str.replace(',', '').str.strip().astype(float)
```

```python
data.info()
```

```python
data['actual_price'] = data['actual_price'].fillna('0')  # Handle NaN values if needed
data['actual_price'] = data['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').str.strip().astype(float)
```

```python
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')  # Convert to float, coerce errors (non-numeric values become NaN)
```

```python
data['rating_count'] = data['rating_count'].fillna('0')  # Handle missing values if necessary
data['rating_count'] = data['rating_count'].str.replace(',', '').str.strip()
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce').fillna(0).astype(int)  # Convert to int and handle errors
```

```python
data['discount_percentage'] = data['discount_percentage'].fillna('0')  # Handle missing values if needed
data['discount_percentage'] = data['discount_percentage'].str.replace('%', '').str.strip()
data['discount_percentage'] = pd.to_numeric(data['discount_percentage'], errors='coerce').fillna(0).astype(int)  # Convert to int and handle errors
```

```python
# doing the text analysis in the review title

review_data= data1['review_title']
review_data
# no working with the regex for getting the good and the bad reviews for  the product 
# List of words you want to match
keywords = 'Good|Satisfied|Excellent|Great|Marvellous|Excellent|Enjoy|Splendid|Essential|Generous|Recommend|Friendly|Impressive|Interesting|Brilliant|Exciting|Fascinating|Expert|Favourite|Ideal'
matching_reviews = review_data.str.contains(keywords, case=False, regex=True)

# just got the false and true value map with the zero and the one 
# 1 - good
# o - not great
numeric_data=matching_reviews.map({True:1,False:0})
data1['review_title_1']=numeric_data
# their is no use of the review_title and review content column in it 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for ratings
plt.figure(figsize=(10, 6))
sns.histplot(data1['rating'], bins=20, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data1['discount_percentage'], y=data1['rating'])
plt.title('Discount Percentage vs Product Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data1['discounted_price'], y=data1['rating'])
plt.title('Discounted Price vs Product Rating')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
total_reviews = data1.groupby('category')['rating_count'].sum()
sns.barplot(x=total_reviews.index, y=total_reviews.values)
plt.title('Total Number of Reviews by Category')
plt.xlabel('Category')
plt.ylabel('Total Reviews')
plt.xticks(rotation=90)  # Rotate category labels if needed
plt.show()
```

```python
plt.figure(figsize=(10, 6))
correlation_matrix = data1[['discount_percentage', 'rating', 'rating_count', 'discounted_price', 'actual_price']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data1['actual_price'], y=data1['discount_percentage'])
plt.title('Discount Percentage vs Actual Price')
plt.xlabel('Actual Price')
plt.ylabel('Discount Percentage')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data1['rating_count'], bins=30, kde=True)
plt.title('Distribution of Rating Counts')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count the number of 1's (positive reviews)
positive_review_count = data1['review_title_1'].sum()

# Count the number of 0's (negative reviews)
negative_review_count = len(data1) - positive_review_count  # Total count minus positive count

print(f"Number of Positive Review Titles (1): {positive_review_count}")
print(f"Number of Negative Review Titles (0): {negative_review_count}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot to visualize the count of positive and negative reviews
plt.figure(figsize=(8, 5))
sns.barplot(x=['Positive (1)', 'Negative (0)'], y=[positive_review_count, negative_review_count], palette='Set2')
plt.title('Count of Positive vs Negative Review Titles')
plt.ylabel('Count')
plt.show()
```
